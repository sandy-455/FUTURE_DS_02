```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("/content/WA_Fn-UseC_-Telco-Customer-Churn.csv")
df
```

```python
#qno 1 Which contract type has the highest customer churn?
df.groupby(["Contract","Churn"]).size()
```

```python
contract_churn = pd.crosstab(df['Contract'], df['Churn'])

contract_churn.plot(kind='bar')

plt.title("Customer Churn by Contract Type")
plt.xlabel("Contract Type")
plt.ylabel("Number of Customers")
plt.show()
```

```python
###The analysis shows that customers with month-to-month contracts have the highest churn rate compared to customers with one-year and two-year contracts. This indicates that customers with short-term commitments are more likely to leave the service
```

```python
## Business Recommendation
#The company should encourage customers to switch to long-term contracts by offering discounts, loyalty rewards, or bundled services. This can help improve customer retention and reduce churn.
```

```python
# qno2 Which internet service type has the highest customer churn?
import pandas as pd
import matplotlib.pyplot as plt

# Calculate churn by internet service
internet_churn = pd.crosstab(df['InternetService'], df['Churn'])

# Plot chart
internet_churn.plot(kind='bar')

plt.title("Customer Churn by Internet Service Type")
plt.xlabel("Internet Service")
plt.ylabel("Number of Customers")

plt.show()
```

```python
##Insight

## The analysis shows that customers using Fiber Optic internet service tend to have higher churn rates compared to customers using DSL or those without internet service. This may indicate issues related to pricing, service expectations, or network performance.
```

```python
#Business Recommendation

#The company should focus on improving fiber internet service quality and customer support. Additionally, offering competitive pricing or promotional plans for fiber internet users could help reduce churn.
```

```python
#qno 3 Do customers with higher monthly charges churn more frequently?
```

```python
import matplotlib.pyplot as plt

# Histogram of monthly charges for churned customers
plt.hist(df[df['Churn'] == "Yes"]['MonthlyCharges'], bins=20)

plt.title("Monthly Charges of Churned Customers")
plt.xlabel("Monthly Charges")
plt.ylabel("Number of Customers")

plt.show()
```

```python
##Insight

##The analysis indicates that customers with higher monthly charges tend to churn more frequently compared to customers with lower monthly charges. This suggests that pricing may play an important role in customer retention
```

```python
##Business Recommendation

#The company should consider introducing flexible pricing plans, loyalty discounts, or bundled service packages to reduce churn among customers with high monthly charges.
```

```python
# qno 4 Does customer tenure affect churn?
```

```python
import matplotlib.pyplot as plt

# Histogram of tenure for churned customers
plt.hist(df[df['Churn'] == "Yes"]['tenure'], bins=20)

plt.title("Tenure Distribution of Churned Customers")
plt.xlabel("Tenure (Months)")
plt.ylabel("Number of Customers")

plt.show()
```

```python
##Insight

#The analysis shows that most churn occurs among customers with shorter tenure, meaning customers who recently joined the service are more likely to leave compared to long-term customers.
```

```python
##Business Recommendation

#The company should focus on improving the early customer experience by providing better onboarding, promotional offers for new users, and proactive customer support during the first few months to increase customer retention.
```

```python
## Qno 5Which payment method has the highest customer churn?
import pandas as pd
import matplotlib.pyplot as plt

# Calculate churn by payment method
payment_churn = pd.crosstab(df['PaymentMethod'], df['Churn'])

# Plot chart
payment_churn.plot(kind='bar')

plt.title("Customer Churn by Payment Method")
plt.xlabel("Payment Method")
plt.ylabel("Number of Customers")

plt.show()
```

```python
##Insight

#The analysis shows that customers using Electronic Check as their payment method tend to have a higher churn rate compared to customers using automatic payment methods such as credit card or bank transfer.
```

```python
##Business Recommendation

#The company should encourage customers to switch to automatic payment methods by offering small incentives such as discounts, reward points, or cashback. Automatic payments can improve convenience and reduce customer churn.
```

```python
## Qno6Does Monthly Charges affect customer churn?
import matplotlib.pyplot as plt

# Monthly charges distribution by churn
plt.hist(df[df['Churn'] == "Yes"]['MonthlyCharges'], bins=20, alpha=0.7, label="Churned")
plt.hist(df[df['Churn'] == "No"]['MonthlyCharges'], bins=20, alpha=0.7, label="Stayed")

plt.title("Monthly Charges vs Customer Churn")
plt.xlabel("Monthly Charges")
plt.ylabel("Number of Customers")
plt.legend()

plt.show()
```

```python
##Insight

#Customers with higher monthly charges tend to churn more frequently compared to customers paying lower monthly charges.
```

```python
#Business Recommendation

"""The company should reduce churn among high-paying customers by offering:
	•	Loyalty discounts
	•	Bundle service packages
	•	Special offers for long-term customers"""
```

```python
#qno7 Do additional services (Online Security, Tech Support, Device Protection) reduce customer churn?
import pandas as pd
import matplotlib.pyplot as plt

# Churn vs Tech Support
tech_support_churn = pd.crosstab(df['TechSupport'], df['Churn'])

tech_support_churn.plot(kind='bar')

plt.title("Churn vs Tech Support")
plt.xlabel("Tech Support")
plt.ylabel("Number of Customers")

plt.show()
```

```python
#Business Recommendation

"""The telecom company should encourage customers to adopt additional services such as:
	•	Tech Support
	•	Online Security
	•	Device Protection"""
```
